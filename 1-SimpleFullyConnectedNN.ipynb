#### import packages

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader
import torchvision.datasets as datasets
import torchvision.transforms as transforms
```

#### Create fully connected network
`class`: what we are trying to detect/classify (if it is dog or not => 2 classes (dog, not dog))
we are using MNIST dataset containing 28*28 pixel images (784 input) which contains 10 classes:
cat, dog, human, .....

```python
class NN(nn.Module):
    def __init__(self, input_size, num_classes):
        super(NN, self).__init__()
        self.fc1 = nn.Linear(input_size, 50)
        self.fc2 = nn.Linear(50, num_classes)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
    
    
model = NN(input_size=784, num_classes=10)
# test model with random data
x = torch.randn(64, 784) # we are not passing images one by one, we pass them as mini batch (64 images at a time)
result = model(x)
print(result.shape)
        
```

## Set device for train

```python
device  = "cuda" if torch.cuda.is_available() else "cpu"
device
```

## Setting hyperparametes

```python
input_size = 784
num_classes = 10
learning_rate = 0.001
batch_size = 64
num_epochs = 1
```

## Load Dataset

```python
train_dataset = datasets.MNIST(root='dataset/', train=True, transform=transforms.ToTensor(), download=True)
train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)

test_dataset = datasets.MNIST(root='dataset/', train=False, transform=transforms.ToTensor(), download=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=True)
```

## initialize Network

```python
model = NN(input_size=input_size, num_classes=num_classes).to(device)
```

## Loss function and optimizer

```python
criterion = nn.CrossEntropyLoss() # loss function
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

## Train the entwork

```python
# each epoch mean the network has seen all images in dataset
for epoch in range(num_epochs):
    for batch_index, (data, targets) in enumerate(train_loader): # loop for each batch
        # Get data to cuda if possible
        data = data.to(device=device)
        targets = targets.to(device=device)
        
        # Get to correct shape
        data = data.reshape(data.shape[0], -1)
        
        # forward
        train_scores = model(data)
        loss = criterion(train_scores, targets)
        
        # backward propagation
        optimizer.zero_grad()
        loss.backward()
        
        # gradien descent or adam stepp
        optimizer.step()
        
```

## Check accuracy on training & test to see how good our model

```python
def check_accuracy(loader, model):
    if loader.dataset.train:
        print("Checking accuracy on training data.")
    else:
        print("Checking accuracy on test data.")
    num_correct = 0
    num_samples = 0
    model.eval()
    
    with torch.no_grad():
        for x, y in loader:
            x = x.to(device=device)
            y = y.to(device=device)
            x = x.reshape(x.shape[0], -1)
            
            scores = model(x)
            _, predictions = scores.max(1) # max of second dimention
            num_correct += (predictions == y).sum()
            num_samples += predictions.size(0)
            
        print(f"Got {num_correct} / {num_samples} with accuracy {float(num_correct)/float(num_samples)*100:.2f}")
        
    model.train()
```

```python
check_accuracy(train_loader, model)
check_accuracy(test_loader, model)
```
